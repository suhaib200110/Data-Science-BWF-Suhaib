```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#Plotting using only matplotlib
x = np.random.normal(0, 100, 50)
y = np.random.normal(0, 100, 50)
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(x, y)
plt.show()
```

```python
df = pd.DataFrame({'x': x, 'y': y})
df.head()
```

```python
# Plotting using seaborn
sns.scatterplot(data=df, x='x', y='y')
plt.show()
```
